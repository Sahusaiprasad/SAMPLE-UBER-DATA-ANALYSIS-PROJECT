```python
df.dropna(inplace = True)
```

```python
df.drop_duplicates(inplace = True)
```

```python
# convert right data types
df['fare_amount'] = df['fare_amount'].astype(float)
df['distance_km'] = df['distance_km'].astype(float)
```

```python
df = df[(df['fare_amount'] > 0 ) & (df['distance_km'] > 0)]
```

```python
df['pickup_time'] = pd.to_datetime(df['pickup_time'])
df['drop_time'] = pd.to_datetime(df['drop_time'])
df
```

```python
#TRIP DURATION IN MINUTES
df['trip_Duration'] = (df [ 'drop_time'] - df [ 'pickup_time' ]).dt.total_seconds() / 60

#extract time features
df['hour'] = df['pickup_time'].dt.hour
df['day'] = df['pickup_time'].dt.day
df['month'] = df['pickup_time'].dt.month
df
```

```python
# remove neagitive or zero values
df = df[df['distance_km'] > 0]
df = df[df['fare_amount'] > 0]
df = df[df['trip_Duration'] > 0]
df
```

```python
#remove un necessary
df = df.drop(['pickup_lat','pickup_lng','drop_lat','drop_lng'])
df
```
